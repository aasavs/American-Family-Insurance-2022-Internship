```python
import pandas as pd
import numpy as np
```

```python
pmt_data = pd.read_csv("Claims Payment Data July 5.csv")
print(len(set(pmt_data["check_number"])))
pmt_data = pmt_data[pmt_data["company"] != "the General"]
pmt_data = pmt_data.drop_duplicates(subset = ["check_number"])
pmt_data
```

```python
def month_counts(months):
    values = []
    for month in months:
        data = pmt_data[pmt_data["issuedMonth"] == month]
        data1 = data
        data = data[data["check_status"] == ("Cleared" or "Voided" or "Issued")]
        paper = len(data[data["check_delivery_method"] == "Central Print"][data["check_delivery_type"] != "Overnight"][data["payment_method"] == "Check"][data["check_status"] != "Voided"]) + len(data[data["check_delivery_method"] == "Local Print"][data["payment_method"] == "Check"][data["check_status"] != "Voided"])
        overnight = len(data[data["check_delivery_type"] == "Overnight"][data["payment_method"] == "Check"])
        void = len(data1[data1["check_status"] == "Voided"])
        eft = len(data[data["payment_method"] == "Electronic funds transfer"])
        cost = "$" + str(paper + (overnight * 56) + (void * 15))  
        values.append([month,paper,overnight,void,eft,cost])
    df =  pd.DataFrame(values, columns=["Month","Paper Check Count","Overnight Check Count", "Voided Check Count", "EFT Payment Count", "Total Cost"]).set_index("Month")
    return df
```

```python
counts = month_counts([5,6])
counts
```
